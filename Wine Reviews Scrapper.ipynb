```python
# importing required libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
# target URL to scrap
url = "https://www.wine.com/list/wine/red-wine/7155-124"
# headers
headers = {
    'User-Agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"
    }
# send request to download the data
response = requests.request("GET", url, headers=headers)
# parse the downloaded data
data = BeautifulSoup(response.text, 'html.parser')
print(data)
```

```python
# find all the sections with specified class nam
cards_data = data.find_all('div', attrs={'class': 'prodItem_wrap'})
# total number of cards
print('Total Number of Cards Found : ', len(cards_data))
# source code of hotel cards
for card in cards_data:
    print(card)
```

```python
# extract the hotel name and price per room
for card in cards_data:
# get the wine name
    Wine_name = card.find('span',attrs={'class':'prodItemInfo_name'})
# get the wine area
    Wine_description = card.find('span', attrs={'class': 'prodItemInfo_originText'})
##get wine price  
    Wine_price = card.find('span', attrs={'class': 'productPrice_price-regWhole'})

##get rating 
    Wine_rating = card.find('span', attrs={'class': 'averageRating_average'})

##get no of ratings 
    ratings_count = card.find('span', attrs={'class': 'averageRating_count'})
    print(Wine_name.text, Wine_description.text, Wine_price.text,Wine_rating.text, ratings_count.text)
```

```python
# create a list to store the data
scraped_data = []
for card in cards_data:
# initialize the dictionary
    card_details = {}
# get the wine name
    Wine_name = card.find('span',attrs={'class':'prodItemInfo_name'})
# get the wine area
    Wine_description = card.find('span', attrs={'class': 'prodItemInfo_originText'})
##get wine price  
    Wine_price = card.find('span', attrs={'class': 'productPrice_price-regWhole'})
##get rating 
    Wine_rating = card.find('span', attrs={'class': 'averageRating_average'})
##get no of ratings 
    ratings_count = card.find('span', attrs={'class': 'averageRating_count'})  
    # add data to the dictionary
    card_details['Wine_name'] = Wine_name.text
    card_details['Wine_description'] = Wine_description.text
    card_details['Wine_price']  =  "$"+Wine_price.text
    card_details['Wine_rating']  =  Wine_rating.text
    card_details['ratings_count']  =  ratings_count.text

# append the scraped data to the list
    scraped_data.append(card_details)
```

```python
# create a data frame from the list of dictionaries
dataFrame = pd.DataFrame.from_dict(scraped_data)
dataFrame.head(20)
# save the scraped data as CSV file
dataFrame.to_csv('Wine_reviews_data.csv', index=False)
```
